```python
# Regex or Regular Expression
import re # re - regular
```

```python
from typing import Pattern
# using match() - The match function checks if the beginning of a string matches a specified pattern. If the pattern is found at the start, it returns a match object; otherwise, it returns None.

text = "Lakshya Web Coder"
Pattern = r"Lakshya"
match = re.match(Pattern,text)

if match:
  print("Match Found - ",match.group())
else:
  print("No Match Found")
```

```python
# using sreach() - The search function scans the entire string to find the first occurrence of a pattern. Unlike match, it doesn't require the pattern to be at the start.

text = "Lakshya Web Coder"
Pattern = r"Coder"
match = re.search(Pattern,text)

if match:
  print("Match Found - ",match.group())
else:
  print("No Match Found")
```

```python
# User Input

text = "Lakshya Web Coder"
text_data = input("Enter search word : ")
Pattern = text_data
match = re.search(Pattern,text)

if match:
  print("Match Found - ",match.group())
else:
  print("No Match Found")
```

```python
# re.ignorecae

text = "Lakshya Web Coder"
Pattern = r"web"
match = re.search(Pattern,text,re.IGNORECASE)

if match:
  print("Match Found - ",match.group())
else:
  print("No Match Found")

# The re.IGNORECASE flag allows r"web" to match "Web" regardless of case.
```
